## Data Cleaning

```python
# Remove empty columns
netflix_data_cleaned = df.drop(columns=['Unnamed: {}' .format(i) for i in range(12, 26)])
netflix_data_cleaned
```

**Remove all unnamed columns to clean our data**

```python
netflix_data_cleaned.shape
```

```python
# Display columns with missing values and their counts
missing_values = netflix_data_cleaned.isnull().sum()
print(missing_values)

```

```python
# Fill missing values in 'director' and 'country' columns with 'Unknown'
netflix_data_cleaned['director'].fillna('Unknown', inplace=True)
netflix_data_cleaned['country'].fillna('Unknown', inplace=True)

# Fill missing values in 'cast' with 'Unknown'
netflix_data_cleaned['cast'].fillna('Unknown', inplace=True)

# Drop rows with missing 'date_added'
netflix_data_cleaned.dropna(subset=['date_added'], inplace=True)

# Fill missing values in 'rating' with 'Not Rated'
netflix_data_cleaned['rating'].fillna('Not Rated', inplace=True)

# Fill missing values in 'duration' with 'Unknown'
netflix_data_cleaned['duration'].fillna('Unknown', inplace=True)
```

```python
# Re-check for missing values
missing_values_after = netflix_data_cleaned.isnull().sum()
print(missing_values_after)
```

```python
# Convert 'date_added' to datetime format, coercing errors to NaT
netflix_data_cleaned['date_added'] = pd.to_datetime(netflix_data_cleaned['date_added'], errors='coerce')

# Drop rows with missing 'date_added' after conversion
netflix_data_cleaned.dropna(subset=['date_added'], inplace=True)

# Extract year and month for further analysis
netflix_data_cleaned['year_added'] = netflix_data_cleaned['date_added'].dt.year
netflix_data_cleaned['month_added'] = netflix_data_cleaned['date_added'].dt.month
```

```python
# Display the first few rows of the cleaned dataframe
print(netflix_data_cleaned.head())

# Re-check for missing values
missing_values_after = netflix_data_cleaned.isnull().sum()
print(missing_values_after)
```

```python
# Summary statistics for numerical columns
summary_stats = netflix_data_cleaned.describe()
#print(summary_stats)

# Summary statistics for categorical columns
categorical_summary = netflix_data_cleaned.describe(include=['O'])
print(categorical_summary)
```

- # Performing Some Exploratory Data Analysis EDA on our Netflix Data.

## 1. No.of Movies vs TV Shows 

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.countplot(data=netflix_data_cleaned, x='type')
plt.title('Number of Movies and TV Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

**Movies are greater in number than the TV Shows**
- Movies **(more than)** twice the no. of TV Shows


- ## You can ignore the below given chart


```python
# Top 10 directors
top_directors = netflix_data_cleaned['director'].value_counts().head(10)
print(top_directors)

# Plot the top directors
plt.figure(figsize=(12, 8))
top_directors.plot(kind='bar')
plt.title('Top 10 Directors on Netflix')
plt.xlabel('Director')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```

## 2. Top 10 Directors by No. of Titles

```python
# Count occurrences of each director
director_counts = netflix_data_cleaned['director'].value_counts()

# Display the top directors
#print(director_counts.head(10))

import matplotlib.pyplot as plt

known_directors = netflix_data_cleaned[netflix_data_cleaned['director'] != 'Unknown']
known_director_counts = known_directors['director'].value_counts()

#get top 10 directors
top_10_directors = known_director_counts.head(10)
print(top_10_directors)

#plot the data now
plt.figure(figsize=(12, 6))
top_10_directors.plot(kind='bar', color='skyblue')
plt.title('Top 10 Directors on Netflix by Number of Titles')
plt.xlabel('Director')
plt.ylabel('No. of Titles')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

## 3. Top 10 Genres on Netflix

```python
# Split the 'listed_in' column into multiple genres
from collections import Counter

genres = netflix_data_cleaned['listed_in'].str.split(',').apply(lambda x: [i.strip() for i in x]).tolist()
flat_genres = [genre for sublist in genres for genre in sublist]
genre_counts = Counter(flat_genres)

# Top 10 genres
top_genres = genre_counts.most_common(10)
print(top_genres)

# Plot the top genres
plt.figure(figsize=(12, 8))
sns.barplot(x=[genre[0] for genre in top_genres], y=[genre[1] for genre in top_genres])
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert 'date_added' to datetime format, coercing errors to NaT
netflix_data_cleaned['date_added'] = pd.to_datetime(netflix_data_cleaned['date_added'], errors='coerce')

# Drop rows with missing 'date_added' after conversion
netflix_data_cleaned.dropna(subset=['date_added'], inplace=True)

# Extract year and month for further analysis
netflix_data_cleaned['year_added'] = netflix_data_cleaned['date_added'].dt.year
netflix_data_cleaned['month_added'] = netflix_data_cleaned['date_added'].dt.month

# Verify the creation of new columns
print(netflix_data_cleaned[['date_added', 'year_added', 'month_added']].head())
```

- ## Ignore this (It's for Self Learning)

```python
# Number of titles added per year
titles_per_year = netflix_data_cleaned['year_added'].value_counts().sort_index()
print(titles_per_year)

# Plot the trend over time
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
titles_per_year.plot(kind='line')
plt.title('Number of Titles Added to Netflix Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()
```

## 4. No of Titles (Movies & TV Shows) added on the site over time

```python
import matplotlib.pyplot as plt

# Filter the data by type
movies = netflix_data_cleaned[netflix_data_cleaned['type'] == 'Movie']
tv_shows = netflix_data_cleaned[netflix_data_cleaned['type'] == 'TV Show']

# Count the number of titles added per year for each type
movies_per_year = movies['year_added'].value_counts().sort_index()
tv_shows_per_year = tv_shows['year_added'].value_counts().sort_index()

# Plot the data
plt.figure(figsize=(12, 8))
plt.plot(movies_per_year.index, movies_per_year.values, label='Movies', marker='o')
plt.plot(tv_shows_per_year.index, tv_shows_per_year.values, label='TV Shows', marker='o')
plt.title('Number of Titles Added to Netflix Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.legend()
plt.grid(True)
plt.show()
```

## 5. Distribution of Ratings

```python
# Distribution of ratings
rating_counts = netflix_data_cleaned['rating'].value_counts()
print(rating_counts)

# Plot the distribution
plt.figure(figsize=(12, 8))
sns.countplot(data=netflix_data_cleaned, x='rating', order=rating_counts.index)
plt.title('Distribution of Ratings on Netflix')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter out "Unknown" and NaN countries
known_countries = netflix_data_cleaned[netflix_data_cleaned['country'] != 'Unknown']
known_countries = known_countries.dropna(subset=['country'])

# Count occurrences of each country
country_counts = known_countries['country'].value_counts()

# Get the top 10 countries
top_10_countries = country_counts.head(10)

# Plot the data
plt.figure(figsize=(12, 8))
top_10_countries.plot(kind='bar', color='pink')
plt.title('Top 10 Countries Producing Content on Netflix')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
